**3.** Построить прямое произведение массивов (все комбинации с каждым элементом). На вход подается двумерный массив

```python
def product(arr1, arr2):
    result = np.zeros((arr1.shape[0] * arr2.shape[0], arr1.shape[1] * arr2.shape[1]))
    return result

arr1 = np.random.randint(5, size=(3, 3))
arr2 = np.random.randint(5, size=(2, 2))
product(arr1, arr2)
```

```python
arr1 = np.random.randint(0,5,(8,3))
arr2 = np.random.randint(0,5,(2,2))

multiple_compare = (arr1[..., np.newaxis, np.newaxis] == arr2)
rows = (multiple_compare.sum(axis=(1,2,3)) >= arr2.shape[1]).nonzero()[0]
arr1, arr2, rows
```

* __Задача 4__: Реализовать кодирование длин серий (Run-length encoding). Для некоторого вектора x необходимо вернуть кортеж из двух векторов одинаковой длины. Первый содержит числа, а второй - сколько раз их нужно повторить.  
 Например, для x = np.array([2, 2, 2, 3, 3, 3, 5]) ответ (np.array([2, 3, 5]), np.array([3, 3, 1])).

```python
def rle(x):
    count = 1
    list_el = []
    list_counts = []
    for i in range(1, len(x)):
        if x[i] == x[i-1]:
            count += 1
        else:
            list_el.append(x[i-1])
            list_counts.append(count)
            count = 1

    list_el.append(x[-1])
    list_counts.append(count)
    return list_el, list_counts
    

x = [2, 2, 3, 4, 3, 3, 2, 5]

rle(x)
```

```python
def rle_numpy(array):
    y = np.r_[True, array[1:] != array[:-1]]
    values = array[y]
    idx = np.nonzero(y)[0]
    lengths = np.diff(np.r_[idx, array.size])
    return values, lengths

arr_1 = np.array([2, 2, 3, 4, 3, 3, 2, 5])
arr_2 = np.array([2, 2, 2, 3, 3, 3, 5])
rle_numpy(arr_1)
#rle_numpy(arr_2)
```

```python
import math

from scipy.spatial.distance import cdist
```

```python
%%time
x = np.random.randint(45, size=(100,2))
y = np.random.randint(115, size=(100,2))
dist = np.zeros((100, 100))
for i in range(len(x)):
    for j in range(i + 1, len(y)):
        dist[i, j] = dist[j, i] = math.sqrt((x[i, 0] - y[i, 0]) ** 2 + (x[i, 1] - y[i, 1]) ** 2)
# dist
```
